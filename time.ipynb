```python
import requests
import pandas as pd
```

```python
url="https://apihub.kma.go.kr/api/typ01/url/kma_sfctm2.php?tm=202211300900&stn=0&help=1&authKey=hU3QNRwcTGmN0DUcHBxpbQ"
```

```python
params = {
    "tm": "202211300900",
    "stn": "0",
    "help": "1",
    "authKey": "hU3QNRwcTGmN0DUcHBxpbQ"
}
response=requests.get(url,params=params) 
print(response.text)
```

```python
from io import StringIO

lines = [l for l in response.text.split('\n') if not l.startswith('#') and l.strip()]
weather_df = pd.read_csv(StringIO('\n'.join(lines)), sep='\s+', header=None)
weather_df
```

```python
weather_df = weather_df[[0, 3, 11, 13, 15]].copy()
weather_df.columns = ['TM', '풍속', '기온', '습도', '강수량']
```
